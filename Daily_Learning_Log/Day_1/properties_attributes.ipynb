<h2 style="color:purple;"><b>Attribute – 1 : shape</b></h2>

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>It returns the no. of rows, columns and how many items of that shape.</b>
</div>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>Example – 1:</b>
</div>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>Understanding the shape of a NumPy array using a simple 2D array.</b>
</div>


```python
_array_ = np.array([[19,23,45],[20,84,51]])
```

```python
print(_array_)
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>This shape returns only row and column values => because we have only one 2D-array.</b>
</div>


```python
print(_array_.shape)
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>Example – 2:</b>
</div>


```python
arr_b = np.array([
                [[19,23],[20,84],[89,56]], 
                [[21,45],[51, 30],[67,15]]
                 ])
```

```python
print(arr_b)
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>This shape returned all the three values i.e no.of items of such shape, rows and columns=> because we have only one 2D-array.</b>
</div>


```python
print(arr_b.shape)
```
