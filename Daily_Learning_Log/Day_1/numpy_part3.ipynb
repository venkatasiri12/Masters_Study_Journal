<h2 style="color:purple;"><b>Memory Computation</b></h2>


```python
import numpy as np
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>We are importing system library</b>
</div>


```python
import sys
```

```python
arr1 = range(100)
arr2 = np.arange(100)
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>Example – 1:</b>
</div>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>
<ul>
<li>
When we multiply the bytes occupied by one element with total no.of elements, it returns the overall memory occupied by the list
</li>
<li>getsizeof() returns the memory occupied by the certain element of a list</li>
</ul></b>
</div>


```python
memory_of_arr1 =  sys.getsizeof(56)*len(arr1)
```

```python
print(memory_of_arr1)
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>Example – 2:</b>
</div>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>
<ul>
<li>
Similarly we are using the numpy attributes to computer the memory occupied by our arrays
</li>
</ul></b>
</div>


```python
memory_of_arr2 =  arr2.itemsize*arr2.size
```

```python
print(memory_of_arr2)
```

<h2 style="color:purple;"><b>Speed Computation</b></h2>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b>Example – 1:</b>
</div>


```python
import time
```

```python
arr3 = range(1000000)
arr4 = range(1000000, 2000000)
```

```python
start_time = time.time()
```

```python
fastness_of_list = [arr3 + arr4 for arr3, arr4 in zip(arr3,arr4)]
```

```python
print(time.time() - start_time)
```
