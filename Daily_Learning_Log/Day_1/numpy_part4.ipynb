<h2 style="color:purple;"><b>Indexing, Slicing and Iteration</b></h2>

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px; font-weight:bold;">
<ul>
  
  <li>Indexing : Accessing single element from an array </li>
  <p></p>
  <li>Slicing : Accessing more than one element from an array</li>
 <p></p>
  <li>Iteration : Sequentially traversing all the elements of an array</li>
 </ul> 

</div>


```python
import numpy as np
```

<h2 style="color:purple;"><b>For 1D Arrays</b></h2>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> Example 1</b>
</div>


```python
arr1 = np.array([23,10,45,88,94,58])
print(arr1)
```

```python
print(arr1[3])
```

```python
print(arr1[:5])
```

```python
print(arr1[-1])
```

<h2 style="color:purple;"><b>For 2D Arrays</b></h2>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> Example 2</b>
</div>


```python
arr2 = np.arange(12,36,2).reshape(4,3)
```

```python
print(arr2)
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> When we apply indexing of 1D arrays, on 2D-array, it returns entire row </b>
</div>


```python
print(arr2[2])
```

```python
print(arr2[3][2])
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> Printing the sub-array of a 2D-array </b>
<ul>
    <li><b>We need to specify the rows, and columns need to be displayed in the output screen </b></li>
  </ul>
</div>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> 
<p>We are printing the subarray <code>[26 28 30 34]</code> from arr2</p></b></div>


```python
print(arr2[2:,1:])
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> 
<p>We are printing the first column <code>[12 18 24 30]</code> from arr2</p></b></div


```python
print(arr2[:,0])
```

```python
arr1
```

```python
for items in arr1:
    print(items)
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> 
<p>If we iterate the 2D-array using a single variable, it displays the array row-wise</p>

</div>


```python
for items in arr2:
    print(items)
```

<h2 style="color:purple;"><b>np.nditer()</b></h2>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> 
<p>np.nditer() : Inorder to print items of an array element-wise</p>

</div>


```python
for items in np.nditer(arr2):
    print(items)
```
