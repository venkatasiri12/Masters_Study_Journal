<h2 style="color:purple;"><b>Method – 1 : max()</b></h2>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> Example 1</b>
</div>


```python
print(arr1.max())
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> Example 2</b>
</div>



```python
print(arr3.max())
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> For 2D-arrays, we can find the max-value 
<ul>
<li>column_wise => axis = 0</li>
<li>Row_wise => axis = 1</li>
</ul></b>
</div>


```python
print(arr3.max(axis = 0))
```

```python
print(arr3.max(axis = 1))
```
