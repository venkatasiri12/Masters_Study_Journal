# Ways of Creating ND-Arrays

#### Initially, we need to install the numpy library, once we have that - we need to import it within our code, inorder to use it in our code

```python
import numpy as np
```

# Method - 1

## np.array()
#### We need to pass the python list as an parameter to the array()
#### This method converts the list into array

```python
#This is the example of 1D-array creation
arr1 = np.array([1,2,3])
```

```python
print(arr1)
```

```python
arr1.ndim
#ndim -> helps us to check the dimensions of the array
```

```python
#This is the example of 2D-array creation
arr2=np.array([[12,24,25],[19,45,24]])
```

```python
print(arr2)
```

```python
arr2.ndim
#ndim -> helps us to check the dimensions of the array
```

```python
#This is the example of 3D-array creation
arr3=np.array([
    [[1,41],[97,15]],
    [[67,89],[48,70]]
    ])
```

```python
print(arr3)
```

```python
arr3.ndim
#ndim -> helps us to check the dimensions of the array
```

#### All the above 1D, 2D and 3D arrays are created using the np.array() method

```python
print(type(arr1))
```

```python
print(type(arr2))
```

```python
print(type(arr3))
```

# Method - 2

## np.zeros()
#### We need to pass the shape of the array (i.e no.of rows and columns) in the form of tuple to the zeros()
#### This method generates the array of given shape 
##### - in which all elements are zero 
##### - all elements are of float dtype

```python
arr4=np.zeros((3,2))
```

```python
print(arr4)
```

# Method - 3

## np.ones()
#### We need to pass the shape of the array (i.e no.of rows and columns) in the form of tuple to the np.ones()
#### This method generates the array of given shape 
##### - This is similar to zero, but here all the elements are ones 
##### - all elements are of float dtype

```python
arr5=np.ones((2,2))
```

```python
arr5
```

# Method - 4

## np.identity()
#### We need to pass the integer value to the np.ones() 
##### -  This method generates the given integer sized identity matrix
##### - all elements are of float dtype

```python
arr6=np.identity(4)
```

```python
print(arr6)
```

# Method - 5

## np.arange()
#### We can pass atmost 3 parameters:
###### - lower_bound
###### - upper_bound (This is not included in our array)
###### - step_value
#### This method generates the array, that contains the integers of given range 


```python
arr7=np.arange(10)
#Here we considered only upper_bound - This will not be included in our output array
```

```python
print(arr7)
```

```python
print(type(arr7))
```

```python
arr8=np.arange(22,29)
print(arr8)
#Here we considered both the lower_bound and upper_bound - upper_bound is not included in our output array
```

```python
arr9=np.arange(45,89,3)
print(arr9)
#Here we considered all the three parameters lower_bound and upper_bound and step_count - upper_bound is not included in our output array
```

# Method - 6

## np.linspace()
#### We need to pass 3 parameters:
###### - lower_bound
###### - upper_bound (This is not included in our array)
###### - no.of equi_distant points
##### - This method generates the array, that contains the elements within the given range 
##### - all elements are of float dtype
##### - when we calculate the difference between any two consecutive points of the array - their distances are equal/same

```python
arr10=np.linspace(0,6,3)
```

```python
print(arr10)
```

# Method - 7

## np.copy()
#### We need to pass array_name as the parameters
##### - This method generates the copy of the original_array

```python
arr11 = np.arange(2,10,2)
print(arr11)
```

```python
arr12 =  np.copy(arr11)
print(arr12)
```
