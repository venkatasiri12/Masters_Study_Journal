<h2 style="color:purple;"><b>Attribute – 1 : np.shape</b></h2>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px; font-weight:bold;">
  
  <p>np.shape : It returns the shape i.e(no.of rows, no.of columns and how many items of such shape)</p>

</div>


<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> Example 1</b>
</div>


```python
arr1 = np.array([91,25,33])
print(arr1)
```

```python
arr1.shape
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> Example 2</b>
</div>


```python
arr2 = np.array([[56],[89],[10],[21]])
print(arr1)
```

```python
arr2.shape
```

<div style="background-color:#ede9fe; color:black; padding:12px; border-left:4px solid purple; border-radius:8px;">
<b> Example 3</b>

```python
arr3 = np.array([[[67,23,90],[89,32,88],[67,10,52]],[[11,54,91],[28,6,44],[22,9,38]]])
print(arr3)
```

```python
print(arr3.shape)
```
